# Defect Analysis

```python
print('Total defects:', df['defect_count'].sum())
print('Average defects per batch:', df['defect_count'].mean())
print('Max defects in a single batch:', df['defect_count'].max())

print('\nDefects by machine:')
print(df.groupby('machine_id')['defect_count'].agg(['sum', 'mean', 'count']).sort_values(by='sum', ascending=False))
```
